```python
import os

def create_folder_structure():
    # Root folder path
    root_folder_path = "/kaggle/working/FYP/Bscan_gen/Output"
    
    # Subfolder names
    subfolders = ["csv", "bscan", "Healthy", "Defect","Cav_only"]
    
    try:
        # Create the root folder if it doesn't exist
        if not os.path.exists(root_folder_path):
            os.makedirs(root_folder_path)
            print(f"Created '{root_folder_path}' folder.")
        
        # Create subfolders inside the root folder
        for subfolder in subfolders:
            subfolder_path = os.path.join(root_folder_path, subfolder)
            if not os.path.exists(subfolder_path):
                os.makedirs(subfolder_path)
                print(f"Created '{subfolder_path}' subfolder.")
            else:
                print(f"'{subfolder_path}' subfolder already exists.")
    except Exception as e:
        print(f"Error occurred: {str(e)}")
        
create_folder_structure()
```

```python
cir_src_only.out
import os
import zipfile
from IPython.display import display, FileLink

def zip_folder(folder_path, zip_name):
    with zipfile.ZipFile(zip_name, "w", zipfile.ZIP_DEFLATED) as zipf:
        for root, dirs, files in os.walk(folder_path):
            for file in files:
                file_path = os.path.join(root, file)
                zipf.write(file_path, os.path.relpath(file_path, folder_path))

output_folder_path = "/kaggle/working/FYP/Bscan_gen/Output"
output_zip_name = "output_folder.zip"
zip_folder(output_folder_path, output_zip_name)

data_folder_path = "/kaggle/working/FYP/Bscan_gen/Data"
data_zip_name = "data_folder.zip"
zip_folder(data_folder_path, data_zip_name)

display(FileLink(output_zip_name))
display(FileLink(data_zip_name))
```
